## Условный оператор

```
a = int(input())

if a > 0:
    print(a)
else:
    print(-1 * a)
```

Условный оператор имеет следующий формат записи:





```
if условие:
    инструкции 1
else:
    инструкции 2
```

В условии мы чаще всего пользуемся операциями больше, меньше и равно.
Так как одинарное равно в python обозначает присваивание, равно в сравнении записывается как ==.

```
if a == 0:
    # a равно 0
    pass

if a != 0:
    # a не равно 0
    pass

if a < 0:
    # a меньше 0
    pass

if a > 0:
    # a больше 0
    pass

if a >= 0:
    # a не меньше (больше или равно) 0
    pass

if a <= 0:
    # a не больше (меньше или равно) 0
    pass

if a + 10 > 0:
    pass
```

```
a = int(input())
b = int(input())

if a * 2 > b:
    print(a)
else:
    print(b)
print("awdawda")
```

## Цикл while и for

Циклы нужно, чтобы повторить одно и то же действие.
Цикл while используется, когда нужно повторять действие до тех пор, пока не будет выполнено какое то условие.

 Например, мы хотим, чтобы пользователь вводил числа до тех пор, пока новое число не будет делиться на 3. И после этого выведем новое число поделенное на 3.

```
a = int(input())

while a % 3 != 0:
    print("Введите число кратное 3")
    a = int(input())

print(a // 3)
```

```
while условие:
    блок инструкций
остальный код
```

```
a = int(input())

summ = 0

while a % 2 == 0:
    a /= 2
    summ += 1

print(summ)
```

**Срезы**

Чтобы изменять строки, нам нужно использовать срезы (slices). Они работают так:

```
string = "Hello, world!"

begin = 2
end = 5

cut = string[begin:end] # берет отрезок строки с элемента с индексом begin (включительно) и до элемента с индексом end (не включительно)

print(cut)
```

Начало и конец можно не указывать. Если не указать начало - срез будет выполнен с начала строки. Если не указать конец - то до самого конца.  

```
string = "Hello, world!"

print(string[4:])
```

```
string = "Hello, world!"

print(string[:4])
```

```
string = "Hello, world!"

print(string[:])
```

Если левая граница больше правой, получится пустая строка.

```
string = "Hello, world!"

print(string[4:2])
```

В срезах можно использовать отрицательную индексацию. Например, хотим отбросить последний символ:

```
string = "Hello, world!"

print(string[:-1])
```

Или взять только последние шесть элементов:

```
string = "Hello, world!"

print(string[-6:])
```

Как третий параметр в срезе можно использовать размер шага K. Т.е. срез будет брать только каждый K-й символ. 

```
string = "Hello, world!"

begin = 0
end = 9
step = 2

cut = string[begin:end:step] # берет отрезок строки с элемента с индексом begin (включительно) и до элемента с индексом end (не включительно)
print(cut)
```

```
string = "Hello, world!"

print(string[::2])
```

Если поставить этот параметр отрицательным, то строка будет выведена в обратном порядке. 

```
string = "Hello, world!"

print(string[::-1])
```

Можем в начале взять срез в правильном порядке, а потом перевернуть. 

```
string = "Hello, world!"

print(string[3:9][::-1])
```

## Списки

Пусть у нас есть задача, в которой нужно обработать не 1, 2, 10 значений, а сразу 1000 или 1000000. При этом задача поставлена так, что нам нужно хранить эти значения.

Списки в python - аналог массивов в других языках программирования. Они позволяют хранить сразу больше количество элементов в одной переменной.

Элементы списка задаются внутри квадрытных скобочек и через запятую.


```
primes = [2, 3, 5, 7, 11, 13]
rainbow = ['Red', 'Orange', 'Yellow', 'Green', 'Blue', 'Indigo', 'Violet']
```

```
a = [4, 5, 6, 2, 3, 1]
```

Теперь чтобы "достать" какой-то элемент из спика, обратиться к нему, нужно указать название спика и номер элемента.

 **Важно:** элементы нумеруются **с нуля** а не с единицы.

```
primes = [2, 3, 5, 7, 11, 13]

a = primes[0]
b = primes[3]

print(a, b)

print(primes[0])
```

В питоне есть трюк, где можно обратиться к элементу с конца списка с помощью отрицательного индекса. Элементы с конца нумеруются с -1.

```
primes = [2, 3, 5, 7, 11, 13]

last = primes[-1]
second_to_last = primes[-2]

print(last, second_to_last)
```

Длину списка, как и длину строки, можно узнать все так же с помощью функции len()

```
rainbow = ['Red', 'Orange', 'Yellow', 'Green', 'Blue', 'Indigo', 'Violet']

print( len(rainbow) )
```

**Как вывести список:**

Можно поэлементно, с помощью цикла for

```
rainbow = ['Red', 'Orange', 'Yellow', 'Green', 'Blue', 'Indigo', 'Violet']

for i in range(len(rainbow)):
    print(rainbow[i])
```

В python функция print() позволяет вывести все элементы списка. Но обрамляет вывод внутри квадратных скобочек, это не то, что нам обычно нужно.

```
print(rainbow)
```

Для этого в python есть специальный трюк. Внутри print() можно поставить звездочку перед названием массива, и элементы будут выведены без скобочек через пробел.

```
rainbow = ['Red', 'Orange', 'Yellow', 'Green', 'Blue', 'Indigo', 'Violet']

print(*rainbow)
```

**Как считать список:**

Если каждый следующий элемент вводится в новой строке, можно изначально задать пустой список, а потом добавлять в него элементы. Это делается с помощью функции append()

```
N = int(input())

arr = []

for i in range(N):
    current = int(input())
    arr.append(current)

print(*arr)
```

Еще раз, чтобы добавить в массив элемент, надо выполнить следующую операцию:

```
# название.append(элемент)

arr = [1, 2, 3]
arr.append(4)

print(arr)
```

Однако так можно считывать только построчно (т.е. каждый элемент идет на новой строчке, они разделены enter-ом). Если все элементы идут в одной строчке, то функция input() считает их сразу и положит в один элемент типа "строка" (string).

Поэтому для таких случаев мы будем использовать специальную конструкцию:

**Считать список из одной строчки:**

```
arr = [int(x) for x in input().split()]

print(arr[0] + arr[1])
```

```
arr = [int(x) for x in input().split()]

for i in range(len(arr)):

    if i % 2 == 0:
        print(arr[i])

```

Как она работает внутри, пока разбираться не будем. Пока ее стоит запомнить. Но важно понимать, что она делает:
если в одной строчке есть только числа, она считывает их все и добавляет их в объявляемый список. 

```
mylist1 = [int(x) for x in input().split()]
mylist2 = [int(x) for x in input().split()]

if (len(mylist1) > len(mylist2)):
    print("Первый длиннее")
elif (len(mylist1) == len(mylist2)):
    print("Они равны по длине")
else:
    print("Второй длиннее")
```

## Многомерные списки

Помимо одномерных (т.е. с одним индексом) бывают многомерные. Список - это просто объект, содержащий в себе другие объекты. Этими объектами могут быть и другие списки.

```
matr = [[1, 2, 3], [4, 5, 6, 7, 8]]
print(matr)
print(matr[1][1])
```

```
x = int(input())

prime = True

for i in range(2, x - 1):
    if x % i == 0:
        prime = False

if x == 1:
    prime = False

print(prime)

# O(x)
```

**Решето Эратосфена**

Теперь хотим найти все простые числа от 1 до n:

Наблюдение: у составного (не простые и не 1, те которые делятся на что-то кроме 1 и себя) числа есть простой делитель.

```
n = int(input())

primes = [1 for i in range(n + 1)]
# индексы от 0 до n включительно

primes[0] = 0
primes[1] = 0

answer = [] # массив с ответом, все простые числа

for cur_prime in range(2, n + 1):
    if primes[cur_prime] == 1:
        
        answer.append(cur_prime)

        for j in range(2 * cur_prime, n + 1, cur_prime): # (откуда, до куда, шаг)
            primes[j] = 0

    


print(*answer)

# алгоритм работает быстро, почти за O(n)
```

Небольшое улучшение:


Идея: если при проверке простого числа $p$ мы еще не зачеркнули какое-то составное число $m$, то оно не меньше, чем $p \cdot p$. Потому что можем представить его как $a \cdot b$, и хотя бы одно из этих чисел меньше чем $p$. Значит, мы уже должны были зачеркнуть число $m$.

```
n = int(input())

primes = [1 for i in range(n + 1)]
# индексы от 0 до n включительно

primes[0] = 0
primes[1] = 0

answer = [] # массив с ответом, все простые числа

for cur_prime in range(2, n + 1):
    if primes[cur_prime] == 0:
        continue # пропускает текущую итерацию цикла

    answer.append(cur_prime)

    for j in range(cur_prime ** 2, n + 1, cur_prime): # (откуда, до куда, шаг)
        primes[j] = 0


print(*answer)
```

## Поиск в глубину (DFS)

```
def dfs(v): # запускается от номера текущей вершины
    been[v] = True

    for u in adl[v]:
        
        if not been[u]: # == True
            dfs(u)

# c++
# void dfs (int v) {}
```

```
n, m = [int(x) for x in input().split()] # кол-во вершин и ребер соответственно

adl = [[] for x in range(n)]

for i in range(m):
    fr, to = [int(x) - 1 for x in input().split()] # сразу вычитаем 1 из считанных чисел
    # fr -= 1 если нумерация с одного
    # to -= 1 если нумерация с одного
    adl[fr].append(to)
    adl[to].append(fr)

start, finish = [int(x) - 1 for x in input().split()]

been = [False for i in range(n)] # used - массив из n нулей

dfs(start)

if been[finish]:
    print("Connected")
else:
    print("Not connected")
```

```
# что такое  for u in lst

for u in lst:

for i in range(len(lst)):
    u = lst[i]

# на с++
for (auto u : lst) {} # lst - вектор
```

**Определение компонент связности:**



1.   Заполняем been не только 0 и 1, а 0 для "не посещена" и 1,2,3..., если посещена и попала в соответствующую компоненту C
2.   Изначально номер компоненты C = 1
3.   Циклом перебираем все вершины (от 1 до n или от 0 до n - 1)
4.   Если очередная вершина еще не посещена, запускаем из нее dfs. По окончании работы добавляем один к номеру текущей компоненты C += 1
5.   В итоге в C будет лежать общее кол-во компонент, а все вершины одной компоненты будут отмечены одним числом в been

```
import turtle
def koch_curve(t, iterations, length):
  if iterations == 0:
    t.forward(length)
  else:
    
    koch_curve(t, iterations - 1, length)
    t.left(60)
    
    koch_curve(t, iterations - 1, length)
    t.right(120)
    
    koch_curve(t, iterations - 1, length)
    t.left(60)
    
    koch_curve(t, iterations - 1, length)
    
t = turtle.Turtle()

t.right(180)
t.penup()

t.forward(200)
t.right(180)
t.pendown()

for i in range(3):
  koch_curve(t, 3, 20)
  t.right(120)
  
turtle.mainloop()
```
